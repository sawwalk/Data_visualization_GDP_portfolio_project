# Life Expectancy Vs GDP project
### Scope  
In this project I will explore the ralationship life expectancy and GDP accross six countries, as well as how these factors have changed over tiem <br>
In this project I will explore the data by:
* Loading and cleaning data.
* Compare each country in terms of it's average GDP and average life expectancy.
* Create plots to visualise how GDP relates to life expectancy.
* Create plots to visualise time series data. 

### Data sources
- GDP Source: [World Bank](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD)
- Life expectancy Source: [World Health Organization](http://apps.who.int/gho/data/node.main.688)

### Importing Moduals and loading data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
#Loading data 
df = pd.read_csv('all_data.csv')

print(df.head())
print(df.columns)
print(df.Country.unique())
```

### Data cleaning, renaming columns

```python
#Replace all instances of United States of America with USA for brevity.
df.replace('United States of America', 'USA', inplace = True)

#Initiates an empty dictionary
new_col_names = {}

#Adds column names as keys and snake_case names as valuee.
for col_name in df.columns:
    new_col_names[col_name] = col_name.replace(' ', '_').replace('(', '').replace(')','') 

print(new_col_names)
#Renames columns using dictionary.
df.rename(columns=new_col_names, inplace = True)

#Observe new column names.
print(df.columns)
```

### Separate data by country
It will be useful for me to analyse trends in each country separately since looking at how GDP compares to life expectancy in one country keeps many variables constant such as geography, culture, climate etc that would otherwise vary. 

```python
#Create a list of country names
countries = df.Country.unique()
print(countries)

#Use string methods to create a list with 'country name_data'
country_variable_names = []
for country in countries:
    country_variable_names.append(f'{country}_data')

print(country_variable_names)

#Creates separate data frames containing the data from each country 
for i in range(len(countries)):
    country_variable_names[i] = df.loc[df.Country == countries[i], :].reset_index(drop=True)
    
#Check that the data frames look as expected
for name in country_variable_names:
    print(name.head(2))
```

### Plot Life expectancy against GDP
We are dealing with two numerical data types so using scatter plots would be approprate. <br>
Here we make a plot for every country.

```python
for name in country_variable_names:
    #Stores the name of country 
    country = name.Country[0]

    #Creates a linear model using statsmodels.api
    model = sm.OLS.from_formula('Life_expectancy_at_birth_years ~ GDP', name)
    results = model.fit()
    #print(results.summary())
    fitted_values = results.predict(name.GDP)

    #Plots data and then plots line form linear model
    plt.scatter(name['GDP'], name['Life_expectancy_at_birth_years'])
    plt.plot(name.GDP, fitted_values, color="black")
    plt.title(f'Life Expectancy vs GDP for {country}')
    plt.ylabel('Life Expectancy')
    plt.xlabel('GDP')
    plt.show()
    plt.clf()
```

### Analyse single country plots
Since each data frame only contains 16 observations the p values for the linear regression models may be inaccurate. However it is still pretty clear
that in all cases higher GDP is associated with higher life expectancy relative to each country. It appears that in each case GDP is positively correlated with life expectancy. <br>
<br>
In particular the scatter plot of China's dataframe stands out in that there appears to be some nonlinear relationship.

### How does life expectancy and GDP change over time in each country?

```python
#Find the range of years that we have data on. 
oldest = str(df.Year.unique().min())
most_recent = str(df.Year.unique().max())

for name in country_variable_names:
    #Stores the name of country 
    country = name.Country[0]
    
    ax1 = plt.subplot()
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Life expectancy at birth (years)')
    ax1.plot(name['Year'], name['Life_expectancy_at_birth_years'], c = 'green')
    ax1.legend(['Life expectancy at birth (years)'])

    ax2 = ax1.twinx()
    ax2.set_ylabel('GDP')
    ax2.plot(name['Year'], name['GDP'], c = 'red')
    ax2.legend(['GDP'], loc = 6)

    plt.title(f'Life Expectancy and GDP from {oldest} to {most_recent} in {country}')
    plt.show()
    plt.clf()
```

We can see that all countries GDP and life expectancy increase over time except for zimbabwe from 2000 to around 2008 in which Zimbabwe experienced hyperinflation and food shortages.

### What country had the most relative ecconomic growth/change in life expectancy?


```python
#This section creates a new data frame whos rows contain the country, it's percent increase in gdp 
percent_increase = []
columns = ['Country', 'gdp_percent_increase', 'LE_percent_increase']
for name in country_variable_names:
    country = name.Country.iloc[0]
    
    gdp2000 = name.GDP.iloc[0]
    gdp2015 = name.GDP.iloc[-1]
    gdp_percent_increase = ((gdp2015 - gdp2000)/gdp2000) * 100
    
    LE_2000 = name.Life_expectancy_at_birth_years.iloc[0]
    LE_2015 = name.Life_expectancy_at_birth_years.iloc[-1]
    LE_percent_increase = ((LE_2015 -LE_2000)/LE_2000) * 100
    
    new_list = [country, gdp_percent_increase, LE_percent_increase]
    percent_increase.append(new_list)

#Create dataframe
percent_increase_df = pd.DataFrame(percent_increase, columns = columns)


#Creates visualisations comparing ecconomic and life expectancy growth from 2000 to 2015
ax1 = plt.subplot(1, 2, 1)
sns.barplot(x = 'Country', y = 'gdp_percent_increase', data = percent_increase_df) 
ax1.set_xticks(range(len(percent_increase_df.Country)))
ax1.set_xticklabels(percent_increase_df.Country, rotation=30)
plt.title('GDP percent increase')
plt.xlabel('Country')
plt.ylabel('gdp percent increase (2000 - 2015)')

ax2 = plt.subplot(1, 2, 2)
sns.barplot(x = 'Country', y = 'LE_percent_increase', data = percent_increase_df)
ax2.set_xticks(range(len(percent_increase_df.Country)))
ax2.set_xticklabels(percent_increase_df.Country, rotation=30)
plt.title('Life expectancy percent increase')
plt.xlabel('Country')
plt.ylabel('LE percent increase (2000 - 2015)')
plt.subplots_adjust(wspace = 0.5, right = 1.3)

plt.show()
```

We can see that China experienced massive ecconomic growth compared to other countries.
Zimbabwe had the greatest percent increase in life expectancy. 

### Comparing GDP and life expectancy across all data sets
Now as we step back and look at the data set as a whole do we find the same sort of linear relationships between GDP and life expectancy as in the previous section?

```python
sns.scatterplot(x = 'GDP', y = 'Life_expectancy_at_birth_years', data = df, hue = 'Country')
plt.title('Collective data set Life Expectancy vs GDP')
plt.ylabel('Life Expectancy')
plt.xlabel('GDP')
plt.show()
```

### Multi country plot analysis
Clearly large differences GDP between countries make the trends in the above plot very hard to see, and overall is not a usefull visulisation. 

There are large variations between wealth in different countries so now the same relationships will be examined between countries.

In order to do this I will calculate the average gdp of each country and rank them from smallest to largest, then plot the average life expectancies. 

```python
#Create empty list
gdp_means = []
# Column names for new dataframe 
columns = ['Country', 'mean_gdp', 'mean_life_expectancy']
for country_data in country_variable_names:
    new_list = [country_data.Country[0], np.mean(country_data.GDP), np.mean(country_data['Life_expectancy_at_birth_years'])]
    gdp_means.append(new_list)

#Create a new dataframe with country its mean gdp and mean life expectancy. 
mean_gdp_df = pd.DataFrame(gdp_means, columns = columns)
#Sorts form by countries by GDP from smallest to largest 
mean_gdp_df = mean_gdp_df.sort_values(by='mean_gdp').reset_index(drop=True)
print(mean_gdp_df)

country_order = list(mean_gdp_df.Country)
```

### Plotting average life expectancy vs average GDP
Below are two ways of visualizing the same data. <br>
- Bar plot (compare averages)
- Box plot (gives us information about the distribution as well as the median)

```python
plt.bar(mean_gdp_df.Country, mean_gdp_df.mean_life_expectancy, )
plt.plot(mean_gdp_df.Country, mean_gdp_df.mean_life_expectancy, color='red')
plt.title('average life expectancy against countries average gdp from smallest to largest')
plt.xlabel('countries ordered from lowest to highest average gdp')
plt.ylabel('average life expectancy')
plt.show()
plt.clf()

sns.boxplot(x = 'Country', y = 'Life_expectancy_at_birth_years', data = df, hue = 'Country', palette = 'Accent', 
           order = country_order)
plt.xlabel('countries ordered from lowest to highest average gdp')
plt.show()
```

The first thing that these data visualisations reveal is that Zimbabwe has significantly lower GDP and life expectancy than all other countries. 

The Zimbabwe data obscures the relationships between the other countries so lets remove zimbabwe from the boxplot to see the relationship between the other countries more clearly.

```python
#Create new dataframe without zimbabwe (used in bar plot)
mean_gdp_df_not_zimbabwe = mean_gdp_df.loc[~(mean_gdp_df.Country == 'Zimbabwe'), :].reset_index(drop=True)
print(mean_gdp_df_not_zimbabwe)

#Create new dataframe without zimbabwe (used in boxplot)
df_not_zimbabwe = df.loc[~(df.Country == 'Zimbabwe'), :]

#Create boxplot
sns.boxplot(x = 'Country', y = 'Life_expectancy_at_birth_years', data = df_not_zimbabwe, hue = 'Country', palette = 'Accent', 
           order = country_order[1:])
#This plots the average life expectancies as black dots
sns.stripplot(x = 'Country', y = 'mean_life_expectancy', data = mean_gdp_df_not_zimbabwe, color='black', size = 5)
plt.title('life expectancy against countries average gdp from smallest to largest')
plt.xlabel('countries ordered from lowest to highest average gdp')
plt.show()
plt.clf()

#Create bar plot
plt.bar(mean_gdp_df_not_zimbabwe.Country, mean_gdp_df_not_zimbabwe.mean_gdp)
plt.title('Differences in average gdp')
plt.xlabel("country")
plt.ylabel('average gdp')
plt.show()
```

It is clear that the same linear relationship between GDP and life expectancy in one country does not hold over multiple countries.
China has the second highest average gdp but the lowest average life expectancy while Chile has the lowest average gdp but the second highest average life expectancy.

In zimbabwe, dgp and life expectancy is very low. One theory may be that gdp is related to life expectancy when it is extreemely low.

```python
#compare the difference in GDP between Chilie and Zimbabwe
Chile_Zimbabwe_gdp_diff = mean_gdp_df[mean_gdp_df.Country == 'Chile'].mean_gdp.item() - mean_gdp_df[mean_gdp_df.Country == 'Zimbabwe'].mean_gdp.item()
print('Chile gdp - Zimbabwe gdp = ', Chile_Zimbabwe_gdp_diff)

#compare the difference in GDP between USA and Chile
USA_Chile_gdp_diff = mean_gdp_df[mean_gdp_df.Country == 'USA'].mean_gdp.item() - mean_gdp_df[mean_gdp_df.Country == 'Chile'].mean_gdp.item()
print('USA gdp - Chile gdp =      ', USA_Chile_gdp_diff)

```
